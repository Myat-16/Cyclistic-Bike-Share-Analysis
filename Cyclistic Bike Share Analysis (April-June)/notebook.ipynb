```python
import pandas as pd
```

### Data Preparation

```python
d1 = pd.read_csv("/content/202104-divvy-tripdata.csv.zip")
d2 = pd.read_csv("/content/202105-divvy-tripdata.csv.zip")
d3 = pd.read_csv("/content/202106-divvy-tripdata.csv.zip")
```

```python
d1.columns == d2.columns
```

```python
d1.info()
```

```python
d3.columns == d2.columns
```

```python
df  = pd.concat([d1, d2, d3])
```

```python
df.head()
```

```python
df.drop(columns=["start_lat","start_lng","end_lat","end_lng"], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['started_at'].values[0].split(" ")
```

```python
# started as day column and started_at as time column for duration
df['started'] = pd.to_datetime(df['started_at'].apply(lambda x: x.split(" ")[0]))
df['started_at'] = pd.to_datetime(df['started_at'])

df['ended']  = pd.to_datetime(df['ended_at'].apply(lambda x: x.split(" ")[0]))
df['ended_at'] = pd.to_datetime(df['ended_at'])

```

```python
df.head()
```

```python
daily_ride  = df.groupby('started', as_index = False).agg({"ride_id":"nunique"})
daily_ride.head()
```

```python
daily_subscription  = pd.pivot_table(df, index="started", columns="member_casual", values="ride_id", aggfunc="nunique")
daily_subscription.reset_index(inplace = True)
daily_subscription.head()
```

```python
daily_data = daily_ride.merge(daily_subscription, on="started",how= 'left')
daily_data.head()
```

```python
daily_data.to_csv("daily_data.csv", index = False)
```

```python
monthly_ride  = df.groupby('started', as_index = False).agg({"ride_id":"nunique"})
monthly_ride.head()
```

```python
monthly_subscription = pd.pivot_table(df, index="started", columns="rideable_type", values="ride_id", aggfunc="nunique")
monthly_subscription.reset_index(inplace = True)
monthly_subscription.head()
```

```python
monthly_data = monthly_ride.merge(monthly_subscription, on="started",how= 'left')
monthly_data.head()
```

```python
monthly_data['month'] = monthly_ride['started'].dt.month
monthly_data.head()
```

```python
monthly_data.to_csv("monthly_data.csv", index = False)
```

```python
weekly_data = df.groupby('started', as_index = False).agg({"ride_id":"nunique"})
weekly_data.head()
```

```python
weekly_data['weekdays'] = weekly_data['started'].dt.day_name()
weekly_data.head()
weekly_data['month'] = weekly_data['started'].dt.month
weekly_data.head()
```

```python
weekly_data.to_csv("weekly_data.csv", index = False)
```
